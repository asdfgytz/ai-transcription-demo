```python
def detect_pauses(audio_segment, min_silence_len=1000, silence_thresh=-40):
    """
    Detects pauses in the audio segment.
    :param audio_segment: The audio segment to analyze.
    :param min_silence_len: Minimum length of silence to consider it a pause.
    :param silence_thresh: Silence threshold in dB.
    :return: List of start and end times of detected pauses.
    """
    return silence.detect_silence(audio_segment, min_silence_len, silence_thresh)

def split_audio_at_pauses(audio_segment, pauses):
    """
    Splits the audio segment at the detected pauses.
    :param audio_segment: The audio segment to split.
    :param pauses: List of start and end times of detected pauses.
    :return: List of audio segments.
    """
    chunks = []
    start = 0
    for pause in pauses:
        end = pause[0]
        chunks.append(audio_segment[start:end])
        start = pause[1]
    chunks.append(audio_segment[start:])
    return chunks
```
